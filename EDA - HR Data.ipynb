#### **HR Analytics - Employees Engagement**
In this EDA project, we will perform HR Analytics focused on Employee Engagement Analysis using a dataset from Kaggle. The goal is to analyze and visualize the data to extract meaningful insights. The role of an HR professional is not as simple as it seems from the outside, as they are involved in recruitment, resolving employee concerns, maintaining a positive work environment, analyzing performance and efficiency, and handling many other responsibilities. Among these, assessing the performance and efficiency of employees is considered one of the most challenging tasks, and its difficulty increases with the number of employees managed. To address this challenge, we are conducting this Exploratory Data Analysis (EDA) project, where different analyses and visualizations will be performed on the Employee Engagement dataset to derive valuable insights.

```python
plt.figure(figsize=(10,15))
ax = sns.boxplot(x='Sex', y='Salary', data=df)
plt.title('Salary Distribution by Sex')


categories = df['Sex'].unique()

for i, category in enumerate(categories):
    salaries = df[df['Sex'] == category]['Salary']
    
    q1 = salaries.quantile(0.25)
    q2 = salaries.quantile(0.50)
    q3 = salaries.quantile(0.75)
    
    iqr = q3 - q1
    lower_fence = q1 - 1.5 * iqr
    upper_fence = q3 + 1.5 * iqr

    ax.text(i, q1, f'Q1: {q1:.0f}', ha='center', va='center', color='black')
    ax.text(i, q2, f'Median: {q2:.0f}', ha='center', va='center', color='blue')
    ax.text(i, q3, f'Q3: {q3:.0f}', ha='center', va='center', color='black')
    ax.text(i, q1 - iqr * 0.6, f'LF: {lower_fence:.0f}', ha='center', va='center', color='red')
    ax.text(i, upper_fence, f'UF: {upper_fence:.0f}', ha='center', va='center', color='red')

plt.show()
```
